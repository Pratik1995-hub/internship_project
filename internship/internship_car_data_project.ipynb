# Create a model to predict the priceUSD columns

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import  train_test_split
from sklearn.model_selection import KFold

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge , Lasso
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost.sklearn import XGBRFRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import math
```

```python
path = r"C:\Users\91969\Desktop\internship Project\cars_price.csv"

data = pd.read_csv(path)
data
```

```python
data.isnull().sum()                                            # Null value present in volume(cm3) , drive_unit , segment
```

```python
# Droping rows where NaN is present inColumns 'Drive unit' and 'Segment'

data.dropna(subset = ['drive_unit'] , axis = 0 , inplace = True)
data.dropna(subset = ['segment'] , axis = 0 , inplace = True)
```

```python
data.isnull().sum()                                      # Now 3 null value in Volume(cm3)
```

```python
# Making Dummy for columns : make , condition , fualtype , color with frequency > 50

dummy_1_column = ["make" , "condition" , "fuel_type" , "color"]
for col in dummy_1_column :
    frequency = data[col].value_counts()
    k = frequency.index[frequency > 50][:-1]
    for cat in k :
        name = col+'_'+cat
        data[name] = (data[col] == cat).astype(int)
    del data[col]
    print(col)
```

```python
# making Dummy for columns : model , transmission , drive_unit , segment with frequency > 100

dummy_2_column = ["model" , "transmission" , "drive_unit" , "segment"]
for col in dummy_2_column :
    frequency = data[col].value_counts()
    k = frequency.index[frequency > 100][:-1]
    for cat in k :
        name = col+'_'+cat
        data[name] = (data[col] == cat).astype(int)
    del data[col]
    print(col)
```

```python
## Droping 1st column 'Unnamed: 0'

data = data.drop('Unnamed: 0' , axis =1 )
```

```python
## Split the data in data_train , data_test with test_size = 0.2

data_train , data_test = train_test_split(data , test_size = 0.2 , random_state =2 )
```

```python
# Replacing NaN value by mean in Data_train

for col in data_train.columns :
    if data_train[col].isnull().sum() > 0 :
        data_train.loc[data_train[col].isnull() , col ] = data_train[col].mean()
```

```python
# Replacing NaN value by mean in Data_test

for col in data_test.columns :
    if data_test[col].isnull().sum() > 0 :
        data_test.loc[data_test[col].isnull() , col ] = data_train[col].mean()
```

```python
x_train = data_train.drop('priceUSD' , axis = 1) 
y_train = data_train["priceUSD"]
```

```python
# Now , To check Performance  on Test

x_test = data_test.drop('priceUSD' , axis = 1)
```

# Ridge ( Linear Regression )

```python
# Making the list of number for lambda values
lambdaa =np.linspace(5,20,100)                                   # Selecting Value for alpha
```

```python
data_train.reset_index(inplace = True , drop = True )            # Rsset the Index in train data
```

```python
x_train = data_train.drop('priceUSD' , axis = 1) 
y_train = data_train["priceUSD"]
```

```python
mae_list = []                                                     # cross validation to find best value of aplha

for a in lambdaa :
    
    ridge = Ridge(fit_intercept= True , alpha = a)
    kf = KFold( n_splits=10)
    val_error = 0
    
    for train , test in kf.split(x_train):
        ridge.fit(x_train.loc[train] , y_train[train])
        p = ridge.predict (x_train.loc[test])
        val_error += mean_absolute_error(y_train[test] , p)
    
    mae_10cv = val_error /10 
    
    print( a, ":" , mae_10cv)
    mae_list.extend([mae_10cv])
        
```

```python
best_alpha = lambdaa[mae_list == min(mae_list)]

print("Alpha with minimum 10cv error is :" , best_alpha)
## Alpha with minimum 10cv error is : [12.42424242]
```

```python
#============================================================================================
```

```python
## appling Ridge

ridge = Ridge(fit_intercept=True , alpha= best_alpha)

ridge.fit(x_train , y_train)
```

```python
# Now , To check Performance  on Test
x_test = data_test.drop('priceUSD' , axis = 1)
```

```python
predicted_test = ridge.predict(x_test)
```

# Boosting 

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.5f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```
